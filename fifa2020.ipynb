```python
import pandas as pd
data_frame = pd.read_csv ('players_20.csv') 
data_frame.shape
data_frame.describe()
data_frame.
```

```python
df = pd.DataFrame(data_frame, columns=['long_name', 'photo', 'age', 'potential'])
df['diference'] = df['potential'] - df['age']
df.sort_values('diference', ascending=False)
```

```python
import seaborn as sns
sns.set()
graph = sns.scatterplot(x='age', y='potential', data=df)
graph
```

```python
from bokeh.plotting import figure,show
from bokeh.models import HoverTool

TOOLTIPS = HoverTool(tooltips=[
    ("index", "$index"),
    ("(Age, Potential)", "($Age, $Potential)"),
    ("Name", "@Name"),]
)
p = figure(title = "Soccer 2020", x_axis_label= 'age', y_axis_label='potential', width=700, height=700, tools=[TOOLTIPS])
p.circle('age', 'potential', size=10, source=df)
show(p)
```
